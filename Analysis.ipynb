```python
# Convert Time of Day to datetime format
df['Time of Day'] = pd.to_datetime(df['Time of Day'], format='%H:%M', errors='coerce')

# Define function to categorize time periods
def get_time_period(t):
    if pd.isnull(t):
        return 'Unknown'
    hour = t.hour
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

# Apply the function
df['Time Period'] = df['Time of Day'].apply(get_time_period)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Time Period', hue='Accident Severity', order=['Morning', 'Afternoon', 'Evening', 'Night'], palette='coolwarm')
plt.title("Accident Severity by Time Period")
plt.xlabel("Time Period")
plt.ylabel("Number of Accidents")
plt.legend(title='Severity')
plt.show()
```
